Question 3: What are the key advantages and limitations of using LLMs in real-world
applications?
- Advantages of LLMs
  - Human-like Text Generation – Produces natural, coherent language.
  - Automation of Tasks – Speeds up writing, summarizing, coding, etc.
  - Multilingual Support – Handles translation and cross-language tasks.

- Limitations of LLMs
  - Hallucinations – Can produce false or misleading information.

  - Dependence on Data Quality – Poor training data → poor model output.
  - Security Vulnerabilities – Can be exploited with prompt attacks.
  - Over-Reliance by Users – Users may trust AI blindly.
  - Ethical & Legal Issues – Copyright, fairness, transparency concerns.

Question 5: Compare and contrast Langchain and LamaIndex. What unique problems
does each solve?
1. Definition
- LangChain
  - A framework designed to build end-to-end LLM applications, connecting language models with tools, APIs, and complex workflows.
- LlamaIndex
  - A framework mainly focused on connecting LLMs to external data, especially for Retrieval-Augmented Generation (RAG) applications.
2. Purpose
- LangChain
  - Multi-step agents
  - Workflow chains
  - Tool-using AI agents
  - Chatbot applications
- LlamaIndex
  - To make it easy to ingest, index, and retrieve external data so LLMs can answer questions based on private or organizational datasets.
3. Key Functional Focus
- LangChain
  - Tool integration
  - Agents & chains
  - API calling
  - Workflow design
  - Memory components
-  LlamaIndex
  - Data ingestion
  - Index creation
  - Query engines
  - Embedding-based retrieval
  - RAG optimization
4. When to Use Which?
- LangChain
  - Complex workflows
  - LLM agents that use tools (browser, calculator, APIs)
  - Stateful conversation memory
  - Multi-model orchestration
-  LlamaIndex
  - Private data retrieval
  - Fast document indexing
  - RAG systems with high accuracy
  - Enterprise data integration
5. Strengths
- LangChain
  - Highly modular
  - Large ecosystem of integrations
  - Excellent for agents
  - Best for workflows and chaining models
  - Rich support for prompt management and memory
- LlamaIndex
  - Very strong document ingestion pipeline
  - Optimized indexing (tree, list, graph indexes)
  - Simple API for building RAG apps
  - Integrates with databases, PDFs, cloud storage
  - Better performance for enterprise data retrieval

Question 7: Integrate Langchain with a third-party API (e.g., weather, news) and show
how responses can be generated via LLMs.

```python
import requests
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
def get_weather(city):
    # Step 1: Convert city → latitude & longitude using geocoding
    geo_url = f"https://geocoding-api.open-meteo.com/v1/search?name={city}"
    geo_res = requests.get(geo_url).json()

    if "results" not in geo_res:
        return None

    lat = geo_res["results"][0]["latitude"]
    lon = geo_res["results"][0]["longitude"]

    # Step 2: Fetch weather using lat & lon
    weather_url = (
        f"https://api.open-meteo.com/v1/forecast?"
        f"latitude={lat}&longitude={lon}&current_weather=true"
    )
    weather_res = requests.get(weather_url).json()

    return weather_res["current_weather"]
```

```python
llm = ChatOpenAI(
    model="gpt-4o-mini",   # or gpt-4o / gpt-3.5-turbo
    temperature=0.2
)
```

```python
prompt = PromptTemplate(
    input_variables=["city", "weather"],
    template="""
You are an AI assistant. Based on the weather data below, explain the
current weather conditions in simple language.

City: {city}
Weather Data: {weather}

Explain in 4–5 lines:
"""
)

parser = StrOutputParser()
```

```python
chain = prompt | llm | parser
```

```python
city = input("Enter city name: ")

weather_data = get_weather(city)

if weather_data:
    response = chain.invoke({
        "city": city,
        "weather": weather_data
    })
    print("\nAI Response:")
    print(response)
else:
    print("Weather data not found for this city.")
```

Question 8: Create a LamaIndex implementation that indexes a local text file and
retrieves answers from it.


```python
from llama_index import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.openai import OpenAI

# 1. Load your local text file
documents = SimpleDirectoryReader("data").load_data()

# 2. Define your LLM (you can change the model if needed)
llm = OpenAI(model="gpt-4o-mini")

# 3. Create index
index = VectorStoreIndex.from_documents(documents, llm=llm)

# 4. Create a query engine
query_engine = index.as_query_engine()

# 5. Ask a question
while True:
    question = input("\nAsk something: ")
    if question.lower() in ("exit", "quit"):
        break

    response = query_engine.query(question)
    print("\nAnswer:", response)
```

Question 9: Demonstrate combining Langchain with LamaIndex to create a simple
document-based Q&A chatbot.

```python
from llama_index import VectorStoreIndex, SimpleDirectoryReader
from llama_index.langchain_helpers.agents import index_tool
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
from langchain_openai import ChatOpenAI


documents = SimpleDirectoryReader("data").load_data()
index = VectorStoreIndex.from_documents(documents)


qa_tool = index_tool.from_index(
    index,
    name="document_qa",
    description="Use this tool to answer questions from the local documents.",
)


llm = ChatOpenAI(model="gpt-4o-mini")

memory = ConversationBufferMemory(memory_key="chat_history")


conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True,
)


print("Chatbot ready! Ask questions about your documents.\n")

while True:
    query = input("You: ")
    if query.lower() in ["exit", "quit"]:
        break


    response = qa_tool.run(query)

    final_answer = conversation.predict(input=f"User asked: {query}\nAnswer using:\n{response}")

    print("Bot:", final_answer)
```

Question 10: A legal firm wants to use AI to summarize large volumes of legal
documents and retrieve relevant information quickly. Propose a solution using
Langchain and LamaIndex, and explain how it would work in practice

```python
import os
from llama_index import SimpleDirectoryReader, VectorStoreIndex
from llama_index.langchain_helpers.agents import index_tool
from langchain.agents import initialize_agent, AgentType
from langchain_openai import ChatOpenAI


print("Loading and indexing legal documents...")


documents = SimpleDirectoryReader("legal_docs").load_data()


index = VectorStoreIndex.from_documents(documents)


legal_search_tool = index_tool.from_index(
    index,
    name="legal_document_search",
    description="Use this tool to search and retrieve relevant parts of legal documents.",
)

print("Index built successfully.\n")


llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.1)


agent = initialize_agent(
    tools=[legal_search_tool],
    llm=llm,
    agent=AgentType.OPENAI_FUNCTIONS,
    verbose=True,
)


print("Legal AI assistant ready.")
print("Ask any question about your documents.")
print("Type 'exit' to quit.\n")

while True:
    query = input("You: ")

    if query.lower() in ["exit", "quit"]:
        break

    response = agent.run(query)
    print("\nAI:", response, "\n")
```
